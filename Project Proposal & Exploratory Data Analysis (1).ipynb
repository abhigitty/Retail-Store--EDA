# Data Source
https://www.kaggle.com/bravehart101/sample-supermarket-dataset/code

The dataset was downloaded from Kaggle and it is renamed to "retailstore".

Variables:

Ship Mode- is the method of shipping the physical goods and products to the intended customers. 

Segment- specifies the customer segment or marketing groups according to their interests and demographics.

Country- is all in the United States of America (USA). 

City- refers to the cities in USA. 

State- is made up of the states in USA.  

Region- indicates the geographic position in the USA continent. 

Category- consists of the named groups of similar and defined physical products. 

Sub-Category- lists the subdivisions with respect to a given category. 

Sales- are the monetary value of transactions between the Superstore and its customers of physical goods in American Dollars (USD$) measurements.

Quantity- records the number of such products in each sales transaction. 

Discount- states the percentage of monetary deduction from the usual product price, and 'Profit' is the financial gains in USD\$ from each sales transaction.

Profit- is the financial gains in USD from each transaction.

```python
Q1= retailstore['Profit'].quantile(0.25)
Q3= retailstore['Profit'].quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit, upper_limit
```

```python
retailstore['Profit'].shape
```

```python
print(retailstore.Profit < (lower_limit)) or (retailstore.Profit > (upper_limit))
```

```python
new_retailstore = retailstore[(retailstore.Profit > lower_limit) & (retailstore.Profit < upper_limit)]
```

```python
new_retailstore.shape
```

```python
sns.boxplot(data=new_retailstore['Profit'])
```

Hence on identifying and trimming the negligible minor outliers present in profit variable, we can say that values lie within expected range.

# Creating Heatmap of the dataset

```python
#correlation between numerical columns using the Heatmap
plt.figure(figsize=(9,7))
correlation=retailstore.corr()
sns.heatmap(correlation,annot=True, cmap='RdYlGn')
plt.show()
```

# Analysing Numerical and categorical variable

```python
#calculating the statewise profit
profit_sum=retailstore.groupby("State")["Profit"].sum()
profit_sum
```

```python
#Barplot for state wise profit 
plt.figure(figsize=(18,12))
sns.barplot(x=profit_sum.index, y=profit_sum.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Profit")
plt.title('State vs Profit Distribution')
plt.show()
```

From the plot, we can see that California has the highest profit followed by New York, Washngton and Michigan while Texas holds for the lowest profit followed by Ohio, Pennsylvania and Illinois.

```python
# State wise Sales of USA
sales_sum=retailstore.groupby("State")["Sales"].sum()
sales_sum
```

```python
plt.figure(figsize=(18,12))
sns.barplot(x=sales_sum.index, y=sales_sum.values)
plt.ylabel("sales")
plt.xticks(rotation=90)
plt.show()
```

1.From the above plot it is clear that California accounts for the highest sales.

2.The top 5 states with highest sales are California, New York,Texas, Washington and Pennsylvania

3.The top 5 states with lowest sales are North Dakota, Maine, West Virginia and South Dakota and Wyoming.

# State wise sales and profit comparision

```python
#Comparing sales and profit state wise
profit_sales=retailstore.groupby("State")["Profit","Sales"].sum()
profit_sales.plot.bar(figsize=(9,5))
plt.show
```

It is observable that Texas, Pennsylvania, Ohio and illinois have high sales but negative profit rate.

# State wise average discount percent

```python
#State wise average discount percent of USA
Avg_discount_percent=retailstore.groupby("State")["Discount"].mean()*100
Avg_discount_percent
```

```python
#Creating barplot for state wise average discount percent of USA
plt.figure(figsize=(15,10))
sns.barplot(x=Avg_discount_percent.index,y=Avg_discount_percent.values)
plt.ylabel("Avg_discount_percent")
plt.xticks(rotation=90)
plt.show()
```

It is clear that Illinois hold for highest average discount percent followed by Texas, Colorado, Arizona and Florida.

# Analysing category wise sales, profit & discount.

```python
# Analysing category wise sales discount and profit
Category_data=retailstore.groupby(["Category"])[['Sales','Discount', 'Profit']].mean()
Category_data
```

```python
#Pie chart of category wise sales discount profit
Category_data.plot.pie(subplots=True, figsize=(30,28), autopct='%0.1f%%', labels= Category_data.index)
plt.show
```

 From the pie-chart above it is clearly evident that :

1.Technology has the more profit(73.1%) due to high sales(49.1%) and low discount(13.2%)

2.Office supplies have average profit(18.9%) due to less sales(12.9%) and minimum discount(15.7%)

3.Furniture accounts for lowest profit(8.1%) due to high discount rate(17.4%) and average sales rate(37.9%)

# Analysing ship mode wise sales, discount & profit

```python
#Analysing ship mode wise sales, discount and profit
shipmode_data=retailstore.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
shipmode_data.plot.pie(subplots=True,figsize=(30,28), autopct='%0.1f%%',labels=shipmode_data.index)
```

From the above Ship mode pie chart in comparision with sales, discount and profit variables it is observable that:

1.Same day holds for highest sales(25.5%)

2.First class accounts for highest discount(16.5%) and maximum Profit(27%)

3.Standard class gained lowest profit(23.3%) due to low sales(24.5%)

4.Second class has profit(25%) higher than that of same day and sales(25.4%) slightly less than same day.

# Analysing sales and profits of sub-categories

```python
# Checking sales and profit of sub-categories
pivot=pd.pivot_table(retailstore, index='Sub-Category', values='Profit')
pivot.plot(kind='bar')
plt.show()
```

Among Sub-Category items, "Copiers" has the highest profits and "Tables" is in loss. 

```python
# Checking the sales of sub-category
pivot=pd.pivot_table(retailstore, index='Sub-Category', values='Sales')
pivot.plot(kind='bar')
plt.show()
```

Among sub-category items, "Copiers and "Machines" have highest sales. Here we see that "Copiers" and "Machines" have highest and second highest sales. In above two graphs "Machines" had second highest sales but due to more discount it has less profit.
"Fastners", "Labels", "Paper" & "Art" account low sales.
